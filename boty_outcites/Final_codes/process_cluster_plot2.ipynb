```python
lst = [960,69, 736, 1321, 2973, 3553]
yes = []
for ind in res2.index:
    if res2['citing_cl'][ind] in lst and res2['referenced_cl'][ind] in lst:
        yes.append(1)
    else:
        yes.append(0)
res2['yes'] = yes
```

```python
r1 = res2[res2['yes'] == 1]
print(len(r1))
print(len(res2))
res2.to_csv("alecensa_top_clusters.csv", index = False)
```
