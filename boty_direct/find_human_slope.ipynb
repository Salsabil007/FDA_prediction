```python
cls = pd.read_csv("features_prediction_2013.csv")
cls = cls[cls.n > 100]
cls = cls.sample(n=100, replace = False)
cls = cls['cluster'].unique()
cl = []
slope = []
for a in cls:
    slope.append(find(2013, a))
    cl.append(a)
    
final = pd.DataFrame()
final['cluster'] = cl
final['slope'] = slope
final['year'] = 2013
final.to_csv("slope_data_2013.csv", index = False)
```
