```python
##Logistic Regression
    
def crossvalidation2(x_train, y_train):
    cvscores = []
    f1score=[]
    Kfold = KFold(n_splits=10, random_state=1, shuffle=True)
    ft = pd.DataFrame()
    for train_index, test_index in Kfold.split(x_train):
        #print(test_index)
        X_train, X_test = x_train[train_index], x_train[test_index]
        Y_train, Y_test = y_train[train_index], y_train[test_index]
        model = LogisticRegression(random_state=0, C = 0.001).fit(X_train,Y_train)
        pre = model.predict(X_test)
        #prey = np.where(pre > 0.5, 1,0)
        f1score.append(f1_score(Y_test, pre , average="macro"))
        
        prob = model.predict_proba(X_test)
        df = pd.DataFrame(prob, columns = ['col_0','col_1'])
        print("len of df ",len(df))
        ft = pd.concat([ft,df])
        #df['col_1'].hist()
        #plt.show()
    print("mean f1 score: ", np.mean(f1score))
    print("length of ft ",len(ft))
    
    fig = plt.figure(figsize=(6.5,6))
    ft['col_1'].hist()
    plt.show()
    fig.savefig('LR.png')


def train(x_train):
    
    print("hdhdhd before ",len(x_train))
  
    y_train = x_train['class']
    
    x_train = x_train.drop(['class','cluster','year','n','pct_dusted_ccn','pct_dusted_rmcl','pct_of_secbiggest_anc',
                           'pct_in_secbiggest_anc','rcr_low','secbiggest_anc',
                           'pct_of_biggest_anc_newish','pct_of_biggest_anc_new','rage','nih','n_clusts_90_anc'
                           ,'pct_in_biggest_anc','biggest_anc','pct_of_secbiggest_anc_newish'
                           ], axis = 1)
    
    #print(x_train.dtypes)
    x_train =  StandardScaler().fit_transform(x_train) 

    y_train = np.array(y_train)
    model = LogisticRegression(random_state=0, C = 0.001).fit(x_train,y_train) ##logistioc
    crossvalidation2(x_train, y_train)
    print(model.intercept_, model.coef_, model.score(x_train, y_train))
    scores = cross_val_score(model, x_train, y_train, cv=10,scoring = "f1")
    print(scores.mean())
    

x_train = pd.read_csv("data_84.csv")
#x_train = x_train[x_train['n'] > 50]
#x_train = x_train[x_train['is_research_article'] > 0.75]
train(x_train)
```
